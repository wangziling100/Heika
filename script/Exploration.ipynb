```python
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
df = pd.read_csv('../data/all_data.csv')
```

```python
df.columns
```

```python
df = df.rename(columns={'Unnamed: 0': 'index'})
df
```

```python
groups = df.groupby(['uid', 'day', 'week', 'is_part', 'exc_num',  'exc_times'])
```

```python

for group in groups:
    index = group[0]
    print("['uid', 'day', 'week', 'is_part', 'exc_num',  'exc_times']: ")
    print(index)
    group_df = pd.DataFrame(group[1])
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='Fx', data=group_df[['index', 'Fx']]).set_title(index)
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='Fy', data=group_df[['index', 'Fy']])
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='Fz', data=group_df[['index', 'Fz']])
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='Mx', data=group_df[['index', 'Mx']])
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='My', data=group_df[['index', 'My']])
    f, ax= plt.subplots(figsize = (14, 10))
    sns.lineplot(x='index', y='Mz', data=group_df[['index', 'Mz']])
```
