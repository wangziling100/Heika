```python
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
import matplotlib.pyplot as plt
% matplotlib inline
```

# Import Data 

```python
df = pd.read_csv('../data/all_data.csv')
```

```python
df = df.rename(columns={'Unnamed: 0': 'index'})
df
```

```python
df.columns
```

# Data Transformation 

```python
df = df.drop(['Force Input', 'Disturbance Force', 'Velocity','Distance'], axis=1)
df['force'] = df['Fx']**2 + df['Fy']**2 + df['Fz']**2
df['force'] = df['force'].pow(1/2)
df['disturbance_force'] = df['Fx.1']**2 + df['Fy.1']**2 + df['Fz.1']**2
df['disturbance_force'] = df['disturbance_force'].pow(1/2)
df['velocity'] = df['Lx']**2 + df['Ly']**2
df['velocity'] = df['velocity'].pow(1/2)
df
df = df.drop(['Fx', 'Fy', 'Fz', 'Mx', 'My', 'Mz','Fx.1', 'Fy.1', 'Fz.1', 
              'Mx.1', 'My.1', 'Mz.1','Lx', 'Ly', 'Lz', 'Ax', 'Ay', 'Az'], axis=1)
df
df.to_csv('../data/transformed_data.csv')
```

```python
df.columns
```

# Data Analysis 

## analyse data by different user

### velocity

#### average velocity of different user 

```python
df_velocity = df.drop(['front', 'left', 'right','force', 'disturbance_force'], axis=1)
```

```python
df_tmp =  df_velocity[['uid', 'velocity']].groupby(['uid']).mean()
df_tmp = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='uid', y='velocity', data=df_tmp).set_title('average velocity')
```

#### velocity in different exercise (uid = 2)

```python
df_tmp = df_velocity[df_velocity['uid']==2]
df_tmp = df_tmp[['velocity','day', 'week', 'exc_num',  'exc_times']].groupby(['day', 'week', 'exc_num',  'exc_times']).mean()
df_tmp
```

```python
df_tmp = df_tmp.reset_index(drop=True)
df_velocity_exc = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='index', y='velocity', data=df_velocity_exc).set_title('average velocity in different exercise')
```

### disdance (front)

#### average distance of different user

```python
df_distance = df.drop(['left', 'right','force', 'disturbance_force', 'velocity'], axis=1)
```

```python
df_tmp =  df_distance[['uid', 'front']].groupby(['uid']).mean()
df_tmp = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='uid', y='front', data=df_tmp).set_title('average distance')
```

#### distance in different exercise (uid = 2)

```python
df_tmp = df_distance[df_distance['uid']==2]
df_tmp = df_tmp[['front','day', 'week', 'exc_num',  'exc_times']].groupby(['day', 'week', 'exc_num',  'exc_times']).mean()
df_tmp
```

```python
df_tmp = df_tmp.reset_index(drop=True)
df_distance_exc = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='index', y='front', data=df_distance_exc).set_title('average distance in different exercise')
```

```python
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='index', y='front', data=df_distance_exc)
sns.lineplot(x='index', y='velocity', data=df_velocity_exc).set_title('average velocity and distance in different exercise')
```

### force 

#### average distance of different user 

```python
df_force = df.drop(['left', 'right','front', 'disturbance_force', 'velocity'], axis=1)
```

```python
df_tmp =  df_force[['uid', 'force']].groupby(['uid']).mean()
df_tmp = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='uid', y='force', data=df_tmp).set_title('average force')
```

#### force in different exercise (uid = 2)

```python
df_tmp = df_force[df_force['uid']==2]
df_tmp = df_tmp[['force','day', 'week', 'exc_num',  'exc_times']].groupby(['day', 'week', 'exc_num',  'exc_times']).mean()
df_tmp
```

```python
df_tmp = df_tmp.reset_index(drop=True)
df_force_exc = df_tmp.reset_index()
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='index', y='force', data=df_force_exc).set_title('average force in different exercise')
```

```python
f, ax= plt.subplots(figsize = (14, 10))
sns.lineplot(x='index', y='front', data=df_distance_exc)
sns.lineplot(x='index', y='velocity', data=df_velocity_exc)
sns.lineplot(x='index', y='force', data=df_force_exc).set_title('average velocity, distance and force in different exercise')
```
